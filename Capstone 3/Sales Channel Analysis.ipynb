```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data


# Hitung total pembelian untuk masing-masing produk
product_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
total_per_product = df[product_cols].sum()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(total_per_product, labels=product_cols, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Proporsi Pembelian per Produk', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List kolom produk
product_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 
                'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Hitung total pembelian per produk
total_per_product = df[product_cols].sum()

# Ubah ke dataframe
product_df = total_per_product.reset_index()
product_df.columns = ['Product', 'TotalSpent']

# --- ðŸ“Š Bar Plot ---
plt.figure(figsize=(10, 6))
sns.barplot(data=product_df, x='Product', y='TotalSpent', palette='viridis')

plt.title('Total Pembelian per Produk (Selama 2 Tahun)', fontsize=16)
plt.xlabel('Produk')
plt.ylabel('Total Pembelian')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert tanggal ke datetime dengan day first
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True, errors='coerce')

# Buat TotalPurchases kalau belum ada
product_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 
                'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['TotalPurchases'] = df[product_cols].sum(axis=1)

# Group pembelian per bulan
monthly_purchases = (
    df.groupby(df['Dt_Customer'].dt.to_period('M'))['TotalPurchases']
    .sum()
    .reset_index()
)
monthly_purchases['Dt_Customer'] = monthly_purchases['Dt_Customer'].dt.to_timestamp()

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(monthly_purchases['Dt_Customer'], monthly_purchases['TotalPurchases'], marker='o', color='teal')
plt.title('Total Pembelian per Bulan (Selama 2 Tahun)', fontsize=16)
plt.xlabel('Bulan')
plt.ylabel('Total Pembelian')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Daftar kolom produk
product_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 
                'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Hitung korelasi antar pembelian produk
correlation_matrix = df[product_cols].corr()

# --- ðŸ”¥ Heatmap Korelasi ---
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, 
            annot=True,           # Tampilkan angka korelasi
            fmt=".2f",            # Dua angka desimal
            cmap="coolwarm",      # Biru (negatif) ke merah (positif)
            linewidths=0.5,       # Garis tipis antar kotak
            square=True)          # Biar kotaknya persegi
plt.title('Heatmap Korelasi Antar Pembelian Produk', fontsize=16)
plt.show()

#Output ini bisa dibuat bundling dengan korelasi produk yang gede misalnya buah sama daging dll
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom channel pembelian & aktivitas
channel_cols = ['NumWebPurchases', 'NumCatalogPurchases', 
                'NumStorePurchases', 'NumWebVisitsMonth']

# Hitung total aktivitas per channel
total_channel = df[channel_cols].sum().reset_index()
total_channel.columns = ['Channel', 'Total']

# --- ðŸ“Š Bar Plot ---
plt.figure(figsize=(10, 6))
sns.barplot(data=total_channel, x='Channel', y='Total', palette='magma')

plt.title('Total Aktivitas Pelanggan per Channel', fontsize=16)
plt.xlabel('Channel')
plt.ylabel('Total Aktivitas')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pilih kolom channel
channel_cols = ['NumWebPurchases', 'NumCatalogPurchases', 
                'NumStorePurchases', 'NumWebVisitsMonth']

# Hitung korelasi antar channel
corr_channel = df[channel_cols].corr()

# --- ðŸ”¥ Heatmap ---
plt.figure(figsize=(8, 6))
sns.heatmap(corr_channel, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Korelasi Antar Channel Pembelian & Kunjungan', fontsize=14)
plt.tight_layout()
plt.show()
```
