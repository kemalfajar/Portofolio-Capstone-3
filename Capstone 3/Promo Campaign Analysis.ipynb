```python
from scipy.stats import shapiro
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pilih kolom-kolom yang ingin diuji
kolom_uji = ['Frequency_Score', 'Monetary_Score', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5','Response']

# Buat dataframe untuk menyimpan hasil uji
hasil_normalitas = {}

for col in kolom_uji:
    stat, p_value = shapiro(df[col])
    hasil_normalitas[col] = {
        'W-Statistic': stat,
        'p-value': p_value,
        'Normal': 'Yes' if p_value > 0.05 else 'No'
    }

normalitas_df = pd.DataFrame(hasil_normalitas).T
```

```python
# Pastikan p-value bertipe float
normalitas_df['p-value'] = normalitas_df['p-value'].astype(float)

# Visualisasi heatmap p-value
plt.figure(figsize=(8, 2))
sns.heatmap(normalitas_df[['p-value']].T, annot=True, cmap='RdYlGn', cbar=False, fmt=".3f")
plt.title('Heatmap Uji Normalitas (Shapiro-Wilk Test)')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Kolom campaign yang ingin dianalisis
campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']

# Hitung total pelanggan per segmen
segment_counts = df['Segment'].value_counts()

# Buat dataframe kosong untuk menyimpan persentase
acceptance_df = pd.DataFrame()

# Hitung persentase untuk masing-masing campaign
for campaign in campaign_cols:
    accepted_per_segment = df.groupby('Segment')[campaign].sum()
    acceptance_rate = (accepted_per_segment / segment_counts) * 100
    acceptance_df[campaign] = acceptance_rate

# Pastikan urutan segmen konsisten
acceptance_df = acceptance_df.fillna(0).sort_index()

# Plot grouped bar chart
segments = acceptance_df.index
x = np.arange(len(segments))  # lokasi x untuk tiap segmen
width = 0.15  # lebar bar

fig, ax = plt.subplots(figsize=(10,6))

for i, campaign in enumerate(campaign_cols):
    ax.bar(x + i * width, acceptance_df[campaign], width, label=campaign)

# Tambahkan label dan styling
ax.set_xlabel('Segment')
ax.set_ylabel('Persentase Acceptance (%)')
ax.set_title('Persentase Acceptance Campaign per Segmen')
ax.set_xticks(x + width * 2)
ax.set_xticklabels(segments)
ax.legend(title='Campaign')
ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Total accepted dari semua campaign
df['TotalAccepted'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1)

# Buat flag: apakah pernah menerima campaign sebelumnya
df['PreviouslyReceived'] = df['TotalAccepted'].apply(lambda x: 'Pernah' if x > 0 else 'Belum Pernah')

# Buat kolom baru untuk Response Campaign terakhir
# Misalnya di data kamu ada kolom 'Response' sebagai hasil dari campaign terakhir
# Jika tidak ada, kamu bisa anggap AcceptedCmp5 sebagai campaign terakhir
df['LastResponse'] = df['Response']  # atau ganti dengan 'AcceptedCmp5' jika tidak ada 'Response'

# Hitung jumlah user di tiap grup (Pernah / Belum Pernah)
group_counts = df.groupby('PreviouslyReceived').size()

# Hitung jumlah yang memberikan response positif di tiap grup
response_counts = df[df['LastResponse'] == 1].groupby('PreviouslyReceived').size()

# Hitung response rate (persentase yang merespons)
response_rate = (response_counts / group_counts * 100).fillna(0)

# Tampilkan hasil
print("Response Rate Berdasarkan History Campaign:")
print(response_rate)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
bars = plt.bar(response_rate.index, response_rate.values, color=['orange', 'green'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height:.1f}%', ha='center')

plt.title("Perbandingan Efektivitas Campaign: Pernah vs Belum Pernah Menerima")
plt.ylabel("Response Rate (%)")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
